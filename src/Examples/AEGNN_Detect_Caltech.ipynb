```python
import torch_geometric.transforms
from torch.nn.functional import mse_loss

from Datasets.ncaltech101 import NCaltech
import torch
from torch.nn import CrossEntropyLoss
from torch.optim import Adam
from Datasets.batching import BatchManager


from Models.CleanAEGNN.GraphRes import GraphRes as AEGNN
from torch_geometric.data import Data as PyGData
from Models.CleanAEGNN.detection import DetectionModel as detectionWrap

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
image_size: tuple[int, int] = NCaltech.get_info().image_size
input_shape: tuple[int, int, int] = (*image_size, 3)


model = detectionWrap(num_classes= 100, img_shape= image_size, device=device)

model = AEGNN(
    input_shape = input_shape,
    kernel_size = 8,
    n = [1, 16, 32, 32, 32, 128, 128, 128],
    pooling_outputs = 128,
    num_outputs = len(NCaltech.get_info().classes),
).to(device)

def transform_graph(graph: PyGData) -> PyGData:
    graph = model.data_transform(
        graph, n_samples = 25000, sampling = True,
        beta =  0.5e-5, radius = 5.0,
        max_neighbors = 32
    ).to(device)
    return graph
```

```python
#Instantiating the ncaltech dataset
ncaltech = NCaltech(
    root=r"C:\Users\jaime\Documents\Maastricht_work\Master\GNNproj",
    transform=transform_graph
)

# Processing the training part of the dataset
ncaltech.process(modes = ["training"])
```

```python
print(f"Training set size: {ncaltech.get_mode_length('training')}")

example_graph = ncaltech.get_mode_data("training", 0)

idx = 100
print(f"Example graph:\n{example_graph}")
print(f"polarity[{idx}]: {example_graph.x[idx, :].tolist()}")
print("pos = [x, y, t]")
print(f"pos[{idx}]: {example_graph.pos[idx, :].tolist()}")
print(f"label: {example_graph.label}")
print(f"bbox: {example_graph.bbox}")
print(f"obj_contour: {example_graph.obj_contour.shape}")
```

```python
training_set = BatchManager(
    dataset=ncaltech,
    batch_size=8,
    mode="training"
)
classes = ncaltech.get_info().classes

cls_to_idx = dict(zip(classes, range(len(classes))))
model.train()
for i in range(1):
    examples = next(training_set)
    reference = torch.tensor([cls_to_idx[cls] for cls in examples.label], dtype=torch.long).to(device)
    out = model(examples)
   
   
```
