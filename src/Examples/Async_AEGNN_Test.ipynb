# Load torch device

```python
import torch_geometric.transforms
from torch.nn.functional import mse_loss

from Datasets.ncaltech101 import NCaltech
import torch
from torch.nn import CrossEntropyLoss
from torch.optim import Adam
from Datasets.batching import BatchManager
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

# Model Initialization

```python
from Models.CleanAEGNN.GraphRes import GraphRes as AEGNN
from torch_geometric.data import Data as PyGData

image_size: tuple[int, int] = NCaltech.get_info().image_size
input_shape: tuple[int, int, int] = (*image_size, 3)

model = AEGNN(
    input_shape = input_shape,
    kernel_size = 8,
    n = [1, 16, 32, 32, 32, 128, 128, 128],
    pooling_outputs = 128,
    num_outputs = len(NCaltech.get_info().classes),
).to(device)

def transform_graph(graph: PyGData) -> PyGData:
    graph = model.data_transform(
        graph, n_samples = 25000, sampling = True,
        beta =  0.5e-5, radius = 5.0,
        max_neighbors = 32
    ).to(device)
    return graph
```

# Dataset Initialization and processing (from the parsed dataset from the aegnn issues thread)

```python
#Instantiating the ncaltech dataset
ncaltech = NCaltech(
    root=r"D:\Uniwersytet\GNNBenchmarking\Datasets\NCaltech",
    transform=transform_graph
)

# Processing the training part of the dataset
ncaltech.process(modes = ["training"])
```

# Asynchronous model test

```python
from Models.AEGNN.aegnn.asyncronous import make_model_asynchronous

test_set = BatchManager(
    dataset=ncaltech,
    batch_size=1,
    mode="training"
)
example = next(test_set)

model.eval()
out = model(example.clone())

new_event = PyGData(x = example.x[-1, :][None, :], pos = example.pos[-1, :][None, :], batch=torch.zeros(1, dtype=torch.long)).to(device)
example.x = example.x[:-1, :]
example.pos = example.pos[:-1, :]
example.batch = example.batch[:-1]
edges1 = example.edge_index[0, :] == example.x.shape[0]
edges2 = example.edge_index[1, :] == example.x.shape[0]
edges = edges1 | edges2
edges = ~edges
example.edge_index = example.edge_index[:, edges]
example.edge_attr = example.edge_attr[edges, :]

example = example.to(device)
new_event = new_event.to(device)
```

```python
edge_attr = torch_geometric.transforms.Cartesian(cat=False, max_value = 10)
model_async = make_model_asynchronous(model, r = 5.0, grid_size = list(image_size), edge_attributes = edge_attr) # model(aegnn) has to be torch.nn.module and r needs to be a float

out_async = model_async.forward(example.clone())

print(f"out - out_async diff: {mse_loss(out, out_async)}")

out_new_event = model_async.forward(new_event.clone())

print(f"out_new_event - out_async diff: {mse_loss(out_new_event, out_async)}")
print(f"out_new_event - out diff: {mse_loss(out_new_event, out)}")
```
