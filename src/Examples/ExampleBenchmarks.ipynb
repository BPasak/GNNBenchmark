Runtime Measurement

```python
exampleModel = ExampleModel()

measureRuntime(exampleModel)

x = torch.ones([1,10])
out = exampleModel(x)

print("Output:")
print(out)
print("Runtime:")
print(exampleModel.runtime_result.getRuntime())

for module_key in exampleModel._modules:
    runtime_result = exampleModel._modules[module_key].runtime_result
    print(module_key)
    if runtime_result is not None:
        print(str(runtime_result.getRuntime()) + " ns")
    else:
        print("No inference happened")
```
