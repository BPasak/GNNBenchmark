Model Instantiation

```python
from Models.EGSST.EGSST import EGSST

egsst = EGSST(
    gcn_count = 3,
    target_size = (300,200),
    detection_head_config = "",
    YOLOX = False,
    Ecnn_flag = False,
    ti_flag = False,
)

def transform_graph(graph):
    graph.x = graph.x[:10000, :]
    graph.pos = graph.pos[:10000, :]
    graph = egsst.data_transform(graph, beta = 0.0001, radius = 3, min_nodes_subgraph = 1000)
    return graph
```

Dataset Initialization and processing (from the parsed dataset from the aegnn issues thread)

```python
from Datasets.ncaltech101 import NCaltech

#Instantiating the ncars dataset
ncars = NCaltech(
    root=r"D:\Uniwersytet\GNNBenchmarking\Datasets\NCaltech",
    transform=transform_graph
)

# Processing the training part of the dataset
ncars.process(modes = ["training"])
```
